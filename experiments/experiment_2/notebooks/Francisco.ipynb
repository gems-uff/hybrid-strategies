```python
import os, sys
sys.path.insert(1, os.path.join(sys.path[0], '..'))
import database
from hover_graph import HoverGraph as Graph
from snowballing.operations import reload, compare_str, load_work_map_all_years
from snowballing.dbmanager import set_attribute
reload()
import pandas as pd
```

```python
df = pd.read_excel("review SB grafico v3.xls")
df
```

```python
import math
def isnan(num):
    if isinstance(num, str):
        return False
    return math.isnan(num)

MAP = {
    "IEEE": "ieee2015",
    "SPRINGER": "springer2015",
    "ENG VILLAGE": "elcompendex2015",
    "WEB OF SCIENCE": "webofscience2015",
    "SCOPUS": "scopus2015",
    "Science Direct": "sciencedirect2015",
    "ACM": "acm2015"
}

ratio = 0.9
result = []

for i, row in df.iterrows():
    top5 = sorted([
        (compare_str(row.iloc[1], w.name), metakey)
        for metakey, w in load_work_map_all_years()
    ], reverse=True, key=lambda x:x[0])[:5]
    
    keys = []
    keys.append(MAP[row.iloc[2]])
    if not isnan(row.iloc[3]):
        keys.append(MAP[row.iloc[3]])
    if not isnan(row.iloc[4]):
        keys.append(MAP[row.iloc[4]])
            
    if top5[0][0] >= ratio:
        print("-", top5[0][1], keys)
        for key in keys:
            set_attribute(top5[0][1], key, "1")
    else:
        result.append([
            row.iloc[1],
            "\n".join(" - ".join(map(str, x)) for x in top5),
            "\n".join(keys)
        ])

result
```
