```python
#EXECUCAO DA SPRINGER COM 65 REGISTROS
import os, sys
import pandas as pd

# Abertura e leitura do arquivo excel
xls = pd.ExcelFile("povoado.xls")

# Leitura da primeira aba da planilha
df = pd.read_excel(xls, "Springer65",  header=None)

# Criação de lista 1 com as linhas da primeira aba da planilha
lista1=[]

for row in df.iterrows():
    index, data = row
    #lista1.append(tuple(data.tolist()))
    lista1.append(tuple(x.lower() for x in data.tolist()))

print("Lista 1: \n")
print (lista1)

# Leitura da segunda aba da planilha
df2 = pd.read_excel(xls, "SemSpringer",  header=None)

# Criação de lista 2 com as linhas da segunda aba da planilha
lista2=[]

for row2 in df2.iterrows():
    index, data2 = row2
    #lista2.append(tuple(data2.tolist()))
    lista2.append(tuple(x.lower() for x in data2.tolist()))
    
print("Lista 2: \n")
print (lista2)


# Método que calcula a interseção entre os objetos da lista
def intersection(lista1, lista2):
    return list(set(lista1) & set(lista2))

print ("\n Lista dos duplicados: \n")
print(intersection(lista1, lista2))
len (intersection(lista1, lista2))
```
